# Análise Base — E-commerce

Este notebook tem como objetivo realizar a exploração inicial e o tratamento básico
dos dados de um e-commerce, garantindo qualidade e entendimento antes das análises
de negócio.

```python
# Importando Bibliotecas
import pandas as pd
```

```python
# Leitura do Dataset
df = pd.read_csv('ecommerce.csv')
df.head()
```

```python
# Estrutura dos dados
df.info()
```

```python
# Tratamento dos dados (convertendo datas)
df['order_date'] = pd.to_datetime(df['order_date'], format='%Y-%m-%d')
```

```python
# Exploração inicial (status dos pedidos)
df['order_status'].value_counts()
```

```python
# Categorias de produtos
df['product_category'].nunique()
```

```python
# Estados dos clientes
df['customer_state'].nunique()
```

```python
# Tipo de pagamento
df['payment_type'].value_counts()
```

# Conclusão

Com base nesta análise inicial, os dados apresentam boa estrutura para análises
